```python
#simulate last few orbits - will take a second
sa = sim[0]
E0 = sa.calculate_energy()
dE = []
t = []

def collision(reb_sim, col):
    reb_sim.contents._status = 5 # causes simulation to stop running and have flag for whether sim stopped due to collision
    return 0

sa.collision = 'direct'
sa.collision_resolve = collision

s = sim[-1]
dt = sim[-1].t-sim[-2].t
for time in np.linspace(s.t,s.t+dt,200):
    s.integrate(time)
    t.append(s.t)
    dE.append(np.abs((s.calculate_energy() - E0)/E0))
    
plt.plot(t,dE,'.')
plt.yscale('log')
```

## Test that epoch doesn't matter for preserving relative phases of planets

```python
import systems.sys_params as sysp
def get_M(e, w, T, P, epoch):
    f = np.pi/2 - w
    E = 2*np.arctan(np.tan(f/2) * np.sqrt((1-e)/(1+e))) #E and f always in same half of ellipse
    M = E - e*np.sin(E)
    return M + (epoch - T)*2*np.pi/P
```

```python
nomass_sys, vaneye_sys, danjh_sys = sysp.get_system_lists()
```

```python
vaneye_sys
```

```python
i=0
orb_params = ["T1","b1","e1","w1","r1/rs","Fs1","T2","b2","e2","w2","r2/rs",
              "Fs2","T3","b3","w3","e3","r3/rs","Fs3","gam1","gam2"]
samples = np.load("systems/data_files/van_eylen/%s/chain_period_eccentricity_finalone.dat.npy"%vaneye_sys[i])
P1 = float(open("systems/data_files/van_eylen/%s/period_%s.01.txt"%(vaneye_sys[i],vaneye_sys[i]), 'r').readlines()[1].split()[0])
P2 = float(open("systems/data_files/van_eylen/%s/period_%s.02.txt"%(vaneye_sys[i],vaneye_sys[i]), 'r').readlines()[1].split()[0])
P3 = float(open("systems/data_files/van_eylen/%s/period_%s.03.txt"%(vaneye_sys[i],vaneye_sys[i]), 'r').readlines()[1].split()[0])
s = samples.shape
data = pd.DataFrame(samples.reshape(s[0]*s[1],s[2]), columns=orb_params)
```

```python
data.head()
```

```python
N=100
id_ = 1

epoch = np.arange(0,1000,1000./N)
s = data.iloc[id_]
d12, d23, d13 = [], [], []
for e in epoch:
    M1 = get_M(s['e1'],s['w1'],s['T1'],P1,e)
    M2 = get_M(s['e2'],s['w2'],s['T2'],P2,e)
    M3 = get_M(s['e3'],s['w3'],s['T3'],P3,e)
    d12.append(M1 - M2)
    d23.append(M2 - M3)
    d13.append(M1 - M3)
    
plt.plot(epoch, d12, '.', label='M1 - M2')
plt.plot(epoch, d23, '.', label='M2 - M3')
plt.plot(epoch, d13, '.', label='M1 - M3')
plt.xlabel('epoch')
plt.ylabel('difference in mean anomaly')
plt.legend(loc='upper left')
plt.savefig('output/MA_vs_Epoch.png')
```
